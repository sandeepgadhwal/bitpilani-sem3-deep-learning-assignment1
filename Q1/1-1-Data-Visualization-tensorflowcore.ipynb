```python
import tensorflow as tf\
    , glob, os, numpy as np, PIL, matplotlib.pyplot as plt
```

## Create Dataset object

```python
train_list = glob.glob('data/imagenette2-160/train/**/*.JPEG')
val_list = glob.glob('data/imagenette2-160/val/**/*.JPEG')
```

```python
train_ds = tf.data.Dataset.from_tensor_slices(train_list)\
.shuffle(len(train_list), reshuffle_each_iteration=False)

val_ds = tf.data.Dataset.from_tensor_slices(val_list)\
.shuffle(len(val_list), reshuffle_each_iteration=False)
```

```python
train_ds
```

```python
class_names = np.array(sorted({os.path.basename(os.path.dirname(x)) for x in train_list}))
print(class_names)
```

```python
img_height = 160
img_width = 160
def open_image(fp):
    # Convert the incoming image in range 0-255 to -1 to 1 range
    # This is standard and helps in better training of a model.
    # Generaly for larger dataset we would like to calculate the stats of the data to normalize but here we can use a standard one
    #
    img = tf.io.read_file(fp)
    # convert the compressed string to a 3D uint8 tensor
    img = tf.image.decode_jpeg(img, channels=3)
    # resize the image to the desired size
    img = tf.image.resize(img, [img_height, img_width])
    img = (tf.cast(img, tf.float32)-255*.5)/(255*.5)
    return img

def get_label(fp):
    # convert the path to a list of path components
    parts = tf.strings.split(fp, os.path.sep)
    # The second to last is the class-directory
    one_hot = parts[-2] == class_names
    # Integer encode the label
    return tf.argmax(tf.cast(one_hot, tf.int8))
```

```python
def open_fp(filepath):
    #
    arr = open_image(filepath)
    lbl = get_label(filepath)
    return arr, lbl
```

```python
open_fp(train_list[5665])
```

```python
# Set `num_parallel_calls` so multiple images are loaded/processed in parallel.
train_ds = train_ds.map(open_fp, num_parallel_calls=None)
val_ds = val_ds.map(open_fp, num_parallel_calls=None)
```

```python
# Test
for image, label in train_ds.take(1):
    print("Image shape: ", image.numpy().shape)
    print("Label: ", label.numpy())
```

```python
def configure_for_performance(ds):
    ds = ds.cache()
    ds = ds.shuffle(buffer_size=1000)
    ds = ds.batch(16)
    ds = ds.prefetch(buffer_size=tf.data.experimental.AUTOTUNE)
    return ds

train_ds = configure_for_performance(train_ds)
val_ds = configure_for_performance(val_ds)
```
